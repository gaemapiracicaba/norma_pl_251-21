# Lê dados

```python
# Lê o arquivo csv com o nome dos municípios
df = pd.read_csv(input_path / 'tab_pl251.csv')

# Deleta Coluna
df = df.drop(['municipio_nome'], axis=1, inplace=False)

# Results
df.info()
df.head()
```

Tabela de agregação!

```python
df_graph = pd.DataFrame(df.groupby('unidade')['id_municipio'].count())
df_graph
```

Faz o gráfico!

```python
df_graph.plot(kind='bar', color='skyblue', title='Mº Municípios por URAE')
```

```python
# Problemas com SSL
df_mun = geo.load_dataset(db='sp', name='tab.municipio_nome')
df_mun.info()
df_mun.head()
```

```python
# Lê o arquivo csv com o nome dos municípios
df_mun = pd.read_csv(
    'https://raw.githubusercontent.com/michelmetran/sp/main/data/tabs/tab_municipio_nome.csv',
    usecols=['id_municipio', 'municipio_nome'],
)
```

```python
# Merge
df = pd.merge(
    df_mun,
    df,
    how='left',
    left_on='id_municipio',
    right_on='id_municipio',
)

# Results
df.head()
```

```python
# Escreve Tabela
df.to_csv(tabs_path / 'tab_municipio_pl251.csv', index=False)
```
